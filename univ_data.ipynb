```python
import numpy as np
import pandas as pd
from tika import parser

data = pd.read_csv('scopus.csv')
```

```python
data = data.sort_values(['Year'],ascending=False)

data
```

```python
data['authors_names_list'] = data['Author full names'].apply(lambda x: x.split(';'))

data['authors_list'] = data['Authors'].apply(lambda x: x.split(';'))
data['affiliations_list'] = data['Authors with affiliations'].apply(lambda x: x.split(';'))
#data['affiliations_list'] = data['Affiliations'].apply(lambda x: x.split(';'))
data['authors_id'] = data['Author(s) ID'].apply(lambda x: x.split(';'))


data['authors_good_names'] = list(list(zip(x, y, z, zz)) for x,y,z,zz in zip(data['authors_list'],data['authors_names_list'],data['affiliations_list'],data['authors_id']))

data = data.explode(column='authors_good_names',ignore_index=True)
#data['author_name']= data.explode(column='authors_good_names',ignore_index=True).apply(lambda x: x)
#data['author_full_name']= data.explode(column='authors_good_names',ignore_index=True).apply(lambda x: x)
#data['author_affiliation']= data.explode(column='authors_good_names',ignore_index=True).apply(lambda x: x)


# data[['author_name','author_full_name','author_affiliation']] = data['authors_good_names'].apply(lambda x: x)

data[['author_name','author_full_name','author_affiliation','author_id']] =pd.DataFrame(data['authors_good_names'].to_list(), index=data.index)



```

```python
data.columns
```

```python
data[['DOI','Title','Link','author_name', 'author_full_name',
       'author_affiliation', 'author_id']].to_csv('paper_authors.csv',index=False)
```

```python
authors_unique_data = data[['author_id','author_name', 'author_full_name','author_affiliation']].drop_duplicates('author_id',keep='first')
```

```python
authors_eloued = authors_unique_data[authors_unique_data['author_affiliation'].str.contains(r'\bEl[ -]Oued\b', case=False, regex=True)].reset_index(drop=True)
authors_eloued.to_csv('authors_eloued.csv',index=False)
```

```python
authors_eloued
```

```python
authors_non_eloued = authors_unique_data[~authors_unique_data['author_affiliation'].str.contains(r'\bEl[ -]Oued\b', case=False, regex=True)].reset_index(drop=True)
authors_non_eloued.to_csv('authors_non_eloued.csv',index=False)
```

```python
authors_non_eloued
```

```python
authors_non_eloued['cleaned_author_name'] = authors_non_eloued['author_full_name'].apply(lambda x : x.split('(')[0].replace(',',''))
cleaned_authors_non_eloued = authors_non_eloued[authors_non_eloued['cleaned_author_name'].apply(lambda x: '.' not in x) ]

cleaned_authors_non_eloued['author_id'] = cleaned_authors_non_eloued['author_id'].str.strip()
cleaned_authors_non_eloued['cleaned_author_name'] = cleaned_authors_non_eloued['cleaned_author_name'].str.strip()


cleaned_authors_non_eloued[['author_id','cleaned_author_name']].to_csv('cleaned_authors_non_eloued.csv',index=False)
```
